## Introduction & Motivation
Over the past few decades, there has been a lot of controversy in the media concerning so called 'tax havens'. Large multinational firms hire accountants from the 'Big-4' accounting companies to think of clever tricks and complicated organizational structures. By doing this, the firms are able to reduce their worldwide tax bill to an absolute minimum. When these multinationals or the accounting firms are asked for a response, their defense typically is that everything they did was legal. They're not wrong, the measures taken are extremely complicated and perfectly legal. The [Luxembourg Leaks](https://www.icij.org/investigations/luxembourg-leaks/ "ICIJ LuxLeaks Homepage") investigation by the International Consortium of Investigative Journalists (ICIJ) showed that these multinationals were able to lower their effective tax rate to 1%, or even lower than that in some cases. 

It isn't only firms that are capable of doing this. More investigations showed that many celebrities, world leaders and criminals found ways to keep their wealth through tax avoidance. The ICIJ uncovered a massive amount of documents called the [Panama Papers](https://panamapapers.icij.org/ "ICIJ Panama Papers Homepage") that revealed all this. In 2017, the ICIJ discovered even more. The [Paradise Papers](https://www.icij.org/investigations/paradise-papers/ "ICIJ Paradise Papers Homepage") revealed offshore interests of more than a hundred firms and elite figures such as Queen Elizabeth and members of Donald Trump's administration. 

As a MSc Economics student in the 'Public Policy' track I find it interesting to what extent firms will go to reduce their tax bills and what these measures look like. I'm particularly interested in if the marginal tax rate is a tool that tax havens use to attract firms or other special rulings and deduction methods. Moreover, how do other countries respond when determining their corporate tax rates to fight in such a competitive environment. In addition to that, is there a change in corporate tax rates set by these 'tax havens' after the revealed documents?

## Model


Let's start of with accessing the excel file I made using the corporate tax rate data from [KPMG](https://home.kpmg.com/xx/en/home/services/tax/tax-tools-and-resources/tax-rates-online/corporate-tax-rates-table.html 'KPMG Corporate Tax Rate Data Source'). The pandas library is used and in the code I show the path it has to take to get to the file on my computer. In addition to that, here is a quick overview of what my data looks like:

```python
tax_data = pd.read_excel('C:/Users/Bastiaanse/Dropbox/Master Economics 2017-2018 UVT/Applied Economic Analysis 1/CorporateTaxRates.xlsx', sheetname='Blad3')
year = tax_data['YEAR'] #Sets the first column in the dataset as the year
countries = tax_data.loc[:,'Albania':'United Kingdom'] #Locates the columns that represent countries and classifies them as that variable
region_averages = tax_data.loc[:,'Africa average':'South America average'] #Locates the columns that represent region averages and classifies them as that variable

tax_data.head(5) #To give an overview of what the data looks like.
```

To use the data and plot it into graphs, I tell Python where to locate the data and classify them as a variable. I did this for the years, individual countries and regions. (The data on region averages didn't specify in what way the average was calculated. It could be a weighted average based on population, or just a sum of corporate tax rates divided by the number of countries.) The timeframe of the time series data is from 2003 to 2017 and contains more than 50 tax jurisdictions or averages of regions.

### Global Overview
The focus of this assignment will be mostly on European corporate tax rates and how they relate to the rest of the world and tax havens. Still, it is important to get a good overview of how tax rates in Europe hold up globally. <a id="Trend"></a>

```python
plt.figure('Average Tax Rates: Regions')
plt.plot(year,region_averages['EU average'])           #Visualization for the overview of EU, European and world wide tax rates.
plt.plot(year,region_averages['Europe average'])
plt.plot(year,region_averages['Global average'])
plt.plot(year,region_averages['OECD average'], 'y-') #I like to use yellow to make a clearer distinction!
plt.xlabel('Year')                                      
plt.ylabel('Corporate Tax Rates')
plt.yticks([20.0, 25.0, 30.0, 35.0], ['20%', '25%', '30%', '35%'])
plt.title('Average Tax Rates: Regions')
plt.legend(['EU Average', 'Europe Average', 'Global Average', 'OECD Average'])
plt.show()
```

 <a id="Race"></a> As we can see in the graph, European tax rates are already much lower than the tax rates around the globe. While that is the European average, the average tax rate for the European Union is even lower. What does stick out, is that there is an obvious trend going on across the globe when it comes to tax competition. The average corporate tax rate is more than five percentage points lower in the plotted regions. This is usually referred to as a ['Race to the Bottom'](https://en.wikipedia.org/wiki/Race_to_the_bottom 'Race to the Bottom Wikipedia')

### Tax Havens
The European Union has made an [official list](http://www.consilium.europa.eu/media/31945/st15429en17.pdf 'EU: List of tax havens') of what they call 'non-cooperative jurisdictions for tax purposes'. At that time, there were seventeen tax jurisdictions on the list. However, there is [criticism](https://www.politico.eu/pro/eu-places-17-countries-on-tax-haven-blacklist-but-none-of-its-own/ 'Criticsm on EU tax haven list') as to why some of the EU member states such as Luxembourg, Ireland and the Netherlands are not on this list. These three countries were known for being favorable to reduce your global tax bill. To take a look at these three countries and their tax rates, see the following graph:

```python
plt.plot(year,countries['Luxembourg'])
plt.plot(year,countries['Ireland'])             #The 3 criticised countries and the global average.
plt.plot(year,countries['Netherlands'])
plt.plot(year,region_averages['Global average'], 'y-') 
plt.xlabel('Year')
plt.ylabel('Corporate Tax Rates')
plt.yticks([10.0, 15.0, 20.0, 25.0, 30.0, 35.0], ['10%', '15%', '20%', '25%', '30%', '35%'])
plt.title('Tax Rates: EU "Tax Havens" vs. Global Average')
plt.legend(['Luxembourg','Ireland','Netherlands','Global Average'])
plt.show()
```

As we can see in the graph, the tax rate of Ireland is much lower (12.5%) than the other two countries in question and the global average. Many large companies are set up in Ireland, such as Google, Facebook and Apple. The country is definitely attractive for business mostly because of the low tax rates. Even when Apple was forced to pay a huge sum of cash, Ireland didn't [want](https://www.nbcnews.com/tech/apple/apple-agrees-pay-15-4-billion-back-taxes-ireland-even-n826701 'Apple & Ireland Tax bills') it. Although the Netherlands and Luxembourg are known as tax havens, they are both above the global average. This goes to show that there are more factors that determine the tax bill than just the corporate tax rate. 

Let's have a look at the two countries that were the big players in the two major tax scandals, LuxLeaks and the Panama Papers:

```python
fig = plt.figure()                                        #Subplotting to have two graphs together.
plt.subplot(2,1,1)                                        #Two rows of plots, one column (To stack them), first plot.
plt.plot(year,countries['Panama'])                        
plt.plot(year,region_averages['Global average'])
plt.plot(year,region_averages['Europe average'])
plt.plot(year,region_averages['OECD average'], 'y-')
plt.axvline(x=2015, color='k', linestyle=':')             #Dotted line to make the distinction better.
plt.title('Panama vs. the World')
plt.legend(['Panama', 'Global Average', 'Europe Average', 'OECD Average'])

plt.subplot(2,1,2)
plt.plot(year,countries['Luxembourg'])
plt.plot(year,region_averages['Global average'])
plt.plot(year,region_averages['Europe average'])
plt.plot(year,region_averages['OECD average'], 'y-')
plt.axvline(x=2014, color='k', linestyle=':')
plt.title('Luxembourg vs. the World')
plt.legend(['Luxembourg', 'Global Average', 'Europe Average', 'OECD Average'])
plt.show()
```

In the first graph we have the tax rates for Panama and a few other tax regions. The graph shows that Panama's corporate tax rate has been well above the average tax rates around the world. It has remained very stable, only seeing one drop from 30% to 25%. Even then, the country doesn't follow the same trend as much as the European Average. The dotted black line represents when the Panama Papers were published. After this happened, there has not been any noticable change in corporate tax rates in both the country or around the world.

The second graph shows the same plot, but then for Luxembourg. Again, the dotted line shows the time all the secret documents were uncovered and in this case that is the Luxleaks. Just like Panama, Luxembourg is far above the corporate tax rates of other regions. Luxembourg hasn't moved towards the global average like Panama did. It has remained fairly stable above it. These two graphs show that the tax havens don't attract firms using the corporate tax rate, if that were the case the opposite would be true: almost no business in these countries. They also show that after all these documents are available to the public, the countries don't seem to change their behavior in terms of the tax rate.

All the 'non-tax havens' also don't change their tax rate setting after these publications. What we do see is the downwards trend of corporate tax rates that I mentioned [earlier](#Race), this 'Race to the bottom' could be caused by the existence of these tax havens. The downward trend could be to fight back against these tax havens attracting their firms to make their own country more attractive.

There must be other ways for tax jurisdictions to attract firms. Of course, there are many. Taxation isn't simply taking a share of the profits, there are many deductibles and exemptions that are available. It's all a matter of how these countries apply different sets of rules to their taxation. The image below shows how countries differ in dealing with dividends.
